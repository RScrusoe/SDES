```python
def my_kw_function(**kwargs):
    if 'name' in kwargs:
        print('Hello ' + str(kwargs['name']))
    if 'course' in kwargs:
        print('Welcome to' + str(kwargs['course']))

my_kw_function(name='Rahul')
my_kw_function(course='AE 425')
```

```python
def add_func_of_integers(func, upto):
    sum = 0
    for n in range(upto+1):
        sum+= func(n)
    return sum
        
def square(n):
    return n**2

print("SUM OF SQUARES UPTO " +str(5) + ' = ' + str( add_func_of_integers(square, 5)))
```

# List comprehensions

```python
n_cubes = [square(i) for i in range(10)]
n_cubes
```

```python
n_cubes = [i*i*i for i in range(10)]
print(n_cubes)
```

```python
N = 6
cartesian_product = [(a,b)
                    for a in range(1,7)
                    for b in range(1,7)]
print(cartesian_product)
print(len(cartesian_product))

odd_sum_cases = [(a,b)
                for a in range(1,7)
                for b in range(1,7)
                if (a + b) % 2 == 1]
print(odd_sum_cases)
```

```python
odd_square = [i**2
             for i in range(15)
             if i%2 == 1]
print(odd_square)
```

# map function

```python
n_numbers = range(10)
list_square = list(map(lambda x:x**2,
                      n_numbers))
print(list_square)

# map function outputs a iterable object and if we iterate thru it, it uses pop function, 
#i.e map object will be gone after iterating thru it once
```

# filter function

```python
even_sum = list(filter(lambda x : (x[0]+x[1]) %2 == 0,
                 cartesian_product))
print(even_sum)
```
